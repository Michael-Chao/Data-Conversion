```python
import os
import pandas as pd
import json

# 定义原始文本文件路径和目标 CSV 文件路径
txt_file_path = r'D:\Project\Convert_Dataset\轨迹预测数据\初步筛选文件夹\20220701.txt'
csv_file_path = r'D:\Project\Convert_Dataset\轨迹预测数据\初步筛选文件夹\file.csv'

# 读取原始文本文件
with open(txt_file_path, 'r') as file:
    lines = file.readlines()

# 跳过前3352行，从第3353行开始处理
data_lines = lines[3352:]

# 初始化 CSV 文件
csv_data = []

for line in data_lines:
    # 将每行数据分割成键值对
    tokens = line.strip().split(', ')
    # 将键值对转换为字典
    data_dict = {}
    for token in tokens:
        key, value = token.split('=')
        # 尝试将值转换为浮点数，如果失败则转换为整数
        try:
            value = float(value)
        except ValueError:
            try:
                value = int(value)
            except ValueError:
                pass  # 如果转换失败，保持 value 为字符串
        # 将键转换为字符串，并用双引号包围
        key = f'"{key}"'
        # 将值添加到字典，不加双引号
        data_dict[key] = value
    
    # 将字典添加到 CSV 数据列表中
    csv_data.append(data_dict)

# 将 CSV 数据列表转换为 DataFrame
df = pd.DataFrame(csv_data)

# 将 DataFrame 保存为 CSV 文件
df.to_csv(csv_file_path, index=False)

# 打印完成消息
print(f"数据已成功转换为 CSV 格式并保存到 '{csv_file_path}'")
```
