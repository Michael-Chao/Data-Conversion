```python
import os
import pandas as pd

# 定义两个文件夹的路径
folder_path_1 = r'D:\Project\Convert_Dataset\轨迹预测数据\0507数据\filter_0507'
folder_path_2 = r'D:\Project\Convert_Dataset\轨迹预测数据\0507数据\original_0507'

# 存储每个文件的数据条目数对比结果
file_row_counts = {}

# 遍历folder_path_1中的所有.csv文件
for file_name in os.listdir(folder_path_1):
    if file_name.endswith('.csv') and os.path.exists(os.path.join(folder_path_2, file_name)):
        # 构造两个文件夹中文件的完整路径
        file_path_1 = os.path.join(folder_path_1, file_name)
        file_path_2 = os.path.join(folder_path_2, file_name)

        # 读取两个文件夹中的CSV文件
        df1 = pd.read_csv(file_path_1)
        df2 = pd.read_csv(file_path_2)

        # 计算数据条目数
        count_1 = len(df1)
        count_2 = len(df2)

        # 打印结果
        print(f'文件 {file_name} 在 filter_0507 文件夹中有 {count_1} 条数据，在 original_0507 文件夹中有 {count_2} 条数据。')

        # 存储结果
        file_row_counts[file_name] = (count_1, count_2)

# 打印所有文件的数据条目数对比结果
for file_name, (count_1, count_2) in file_row_counts.items():
    print(f'文件 {file_name} 在 filter_0507 文件夹中有 {count_1} 条数据，在 original_0507 文件夹中有 {count_2} 条数据。')
```

```python
import os
import json

# 定义文件夹路径
folder_path = r'D:\Project\Convert_Dataset\轨迹预测数据\filter_ADSB_0506'

# 遍历文件夹中的所有.txt文件
for file_name in os.listdir(folder_path):
    if file_name.endswith('.txt'):
        # 构造完整的文件路径
        file_path = os.path.join(folder_path, file_name)
        
        # 尝试打开并读取文件
        try:
            with open(file_path, 'r') as file:
                # 整个文件是一个JSON数组，我们一次性加载它
                data = json.load(file)
                
                # 计算数据条目数
                data_count = len(data)
                
                # 打印结果
                print(f'文件 {file_name} 有 {data_count} 条数据。')
        
        except FileNotFoundError:
            print(f"文件 {file_path} 未找到。")
        except json.JSONDecodeError:
            print(f"文件 {file_name} 的内容不是有效的JSON格式。")
        except Exception as e:
            print(f"读取文件 {file_name} 时发生错误: {e}")
```

```python
import os
import pandas as pd

# 定义文件夹路径
folder_path = r'D:\Project\Convert_Dataset\轨迹预测数据\0507数据\2_original'

# 初始化计数器
count_greater_89_526 = 0
count_less_89_507 = 0

# 遍历文件夹中的所有.csv文件
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        # 构造完整的文件路径
        file_path = os.path.join(folder_path, file_name)
        
        # 读取CSV文件
        df = pd.read_csv(file_path)
        
        # 假设CSV文件中有以下列名'LON'包含经度数据
        # 筛选出经度大于89.526的数据数量
        greater_89_526 = df[df['LON'] > 89.527]
        
        # 筛选出经度小于89.507的数据数量
        less_89_507 = df[df['LON'] < 89.5034]
        
        # 更新计数器
        count_greater_89_526 += len(greater_89_526)
        count_less_89_507 += len(less_89_507)
        
        print(f'文件 {file_name} 处理完成：')
        print(f'  经度大于89.527的数目: {len(greater_89_526)}')
        print(f'  经度小于89.5034的数目: {len(less_89_507)}')

# 打印总数
print(f'所有文件中，经度大于89.52696的总数目: {count_greater_89_526}')
print(f'所有文件中，经度小于89.507的总数目: {count_less_89_507}')
```

```python
'''
处理前后数据点的数目对比
'''
import os
import pandas as pd

# 定义原始文件夹和处理后文件夹的路径
original_folder_path = r'D:\Project\Convert_Dataset\轨迹预测数据\0507数据\2_original'
processed_folder_path = r'D:\Project\Convert_Dataset\轨迹预测数据\0507数据\filter_0507'

# 存储每个文件原始和处理后的行数以及差值
file_row_counts = {}

# 遍历原始文件夹中的所有.csv文件
for file_name in os.listdir(original_folder_path):
    if file_name.endswith('.csv'):
        # 构造原始文件的完整路径
        original_file_path = os.path.join(original_folder_path, file_name)
        
        # 构造处理后文件的完整路径
        processed_file_path = os.path.join(processed_folder_path, file_name)
        
        # 读取原始CSV文件
        original_df = pd.read_csv(original_file_path)
        original_count = len(original_df)
        
        # 读取处理后的CSV文件
        processed_df = pd.read_csv(processed_file_path) if os.path.exists(processed_file_path) else pd.DataFrame()
        processed_count = len(processed_df)
        
        # 计算差值
        difference = original_count - processed_count
        
        # 打印每个文件的原始数据条目数、处理后数据条目数以及处理的数据条目数
        print(f'文件 {file_name} 原始数据条目数: {original_count}, 处理后数据条目数: {processed_count}, 处理的数据条目数: {difference}')
        
        # 存储结果
        file_row_counts[file_name] = (original_count, processed_count, difference)

# 打印所有文件的比较结果
for file_name, (original_count, processed_count, difference) in file_row_counts.items():
    print(f'文件 {file_name} 原始数据条目数: {original_count}, 处理后数据条目数: {processed_count}, 处理的数据条目数: {difference}')
```

```python
import os
import json
import pandas as pd

# 定义三个文件夹的路径
folder_paths = [
    r'D:\Project\Convert_Dataset\轨迹预测数据\filter_ADSB_0506',  # 原始数据
    r'D:\Project\Convert_Dataset\轨迹预测数据\0507数据\original_0507',  # 提取后数据
    r'D:\Project\Convert_Dataset\轨迹预测数据\0507数据\filter_0507'   # 处理后数据
]

# 存储每个文件的数据条目数
file_data_counts = {}

# 遍历第一个文件夹中的所有文件
for file_name in os.listdir(folder_paths[0]):
    if file_name.endswith('.txt'):
        # 构造完整的txt文件路径
        txt_file_path = os.path.join(folder_paths[0], file_name)
        
        # 从第一个文件夹中读取json格式数据
        with open(txt_file_path, 'r') as f:
            data = json.load(f)
            data_count = len(data)
        
        # 初始化文件的数据条目数为0
        file_data_counts[file_name] = [data_count, 0, 0]

# 遍历后两个文件夹中的所有同名csv文件，并更新数据条目数
for folder_path in folder_paths[1:]:
    for file_name in file_data_counts.keys():
        # 构造完整的csv文件路径
        csv_file_path = os.path.join(folder_path, file_name[:-4] + '.csv')
        
        # 读取csv文件并更新数据条目数
        data_count = len(pd.read_csv(csv_file_path))
        file_data_counts[file_name][folder_paths.index(folder_path)] = data_count

# 打印每个文件的数据条目数对比结果
for file_name, counts in file_data_counts.items():
    print(f'文件 {file_name[:-4]} 原始数据条目数: {counts[0]}, 提取后数据条目数: {counts[1]}, 处理后数据条目数: {counts[2]}')

print("所有文件的数据条目数对比完成。")
```
