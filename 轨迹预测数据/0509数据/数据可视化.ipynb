```python
import matplotlib.pyplot as plt

# 读取文件并提取经纬度数据
file_path = r'D:\Project\20240502\降落\out2\20220701_1\4.txt'
lon = []
lat = []
with open(file_path, 'r') as file:
    for line in file:
        values = line.strip().split(',')
        lon.append(float(values[1]))  # 第二列为经度
        lat.append(float(values[2]))  # 第三列为纬度

# 绘制散点图
plt.figure(figsize=(8, 6))
plt.scatter(lon, lat, s=10, c='blue', alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Scatter Plot of Longitude and Latitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 文件路径
file_path = r'D:\Project\20240502\降落\train\0\259.npy'

# 加载.npy文件
data = np.load(file_path)

# 提取经度和纬度数据
lons = data[:, 0]  # 提取第一列作为经度
lats = data[:, 1]  # 提取第二列作为纬度

# 绘制经纬度图形
plt.figure(figsize=(8, 6))
plt.plot(lons, lats, marker='o', linestyle='-')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Trajectory')
plt.grid(True)
plt.show()
```
